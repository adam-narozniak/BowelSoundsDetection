```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()
```

```python
x = np.linspace(-10, 10, 1000)
y = 1 / (1 + np.exp(-x) )

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.rc('xlabel', labelsize=16)
plt.rc('ylabel', labelsize=16)

plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title


plt.figure(figsize=(16, 9))
plt.xlabel("x", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, y , linewidth=4)
plt.title("Sigmoid(x)", fontsize=32)
plt.savefig("./sigmoid.jpg", dpi=300)
```

```python
y= np.array([0 if a <0 else a for a in x])

plt.figure(figsize=(16, 9))
plt.xlabel("x", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, y , linewidth=4)
plt.title("ReLU(x)", fontsize=32)
plt.savefig("./relu.jpg", dpi=300)
```
